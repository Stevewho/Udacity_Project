## data type ##

1. show data type 
2. list feature name that belong to its data type 

```python
#meta data type
data = []
for f in train.columns:
    # Defining the role
    if f == 'target':
        role = 'target'
    elif f == 'id':
        role = 'id'
    else:
        role = 'input'
         
    # Defining the level
    if 'bin' in f or f == 'target':
        level = 'binary'
    elif 'cat' in f or f == 'id':
        level = 'nominal'
    elif train[f].dtype == float:
        level = 'interval'
    elif train[f].dtype == int:
        level = 'ordinal'
        
    # Initialize keep to True for all variables except for id
    keep = True
    if f == 'id':
        keep = False
    
    # Defining the data type 
    dtype = train[f].dtype
    
    # Creating a Dict that contains all the metadata for the variable
    f_dict = {
        'varname': f,
        'role': role,
        'level': level,
        'keep': keep,
        'dtype': dtype
    }
    data.append(f_dict)
    
meta = pd.DataFrame(data, columns=['varname', 'role', 'level', 'keep', 'dtype'])
meta.set_index('varname', inplace=True)



```

```python
# display( pd.concat([meta.head(5),meta.tail(5)]))
display(meta)

levels= meta['level'].unique()
# for level in levels:
#     varname = meta.loc[(meta['level']==level) & meta['keep']].index.tolist()
# #     varname = meta[(meta.level==level)& (meta.keep)].index.tolist()
#     print('\n{0} columns:\n{1}'.format(level,varname))

#manually print
# print('\ninterval:\n ',meta.loc[meta['level']=='interval'].index.tolist())
# print('\nnominal:\n ', meta.loc[meta['level']=='binary'].index.tolist())

writer = pd.ExcelWriter('Variable.xlsx', engine='xlsxwriter')
meta.to_excel(writer, sheet_name='Sheet1')

writer.save()

```

```python
import seaborn as sns

nominals = meta[(meta.level == 'nominal') & (meta.keep)].index

# len(nominals)
# for i in range(len(nomina)):
#     plt.subplot(2, 3, i)
#     plt.text(0.5, 0.5, str((2, 3, i)),
#              fontsize=18, ha='center')
fig=plt.figure(figsize=(15, 30))
# fig, ax = plt.subplots(7,2 ,figsize=(10,30))
fig.subplots_adjust(hspace=0.3, wspace=0.4)
for index, nominal in enumerate(nominals):
    
    
    # Calculate the percentage of target=1 per category value
    cat_perc = train[[nominal, 'target']].groupby([nominal],as_index=False).mean()
    cat_perc.sort_values(by='target', ascending=False, inplace=True)
    # Bar plot
    # Order the bars descending on target mean
#     ax = plt.subplot(221)
    ax = fig.add_subplot(7,2,index+1)
    sns.barplot(ax=ax, x=nominal, y='target', data=cat_perc, order=cat_perc[nominal])
    plt.ylabel('% target', fontsize=18)
    plt.xlabel(nominal, fontsize=18)
    plt.tick_params(axis='both', which='major', labelsize=18)
plt.tight_layout()
plt.show()
# plt.savefig('cat.jpg')
 
```

 ## Binary features

```python
py.init_notebook_mode(connected=True)
from collections import defaultdict 
bin_counts = defaultdict(list)

for col in bincols:
    bin_counts[col]= train[col].value_counts()
    
    
#return a list of numbers of true counts for each coloumn    
y_true = [ bin_counts[col][1] for col in bincols]
y_false = [ bin_counts[col][0] for col in bincols]

trace1 = go.Bar(
    x=bincols,
    y=y_true,
    name='True',
    marker = dict(color='rgb(158,202,225)',
                  line=dict(color='rgb(8,48,107)',width=0.1)
                 )
)

trace2 = go.Bar(
    x=bincols,
    y=y_false,
    name='False',
    marker = dict(color='rgb(222,190,200)',
                  line=dict(color='rgb(8,48,107)',width=0.1)
                 )
)

data = [trace1, trace2]
layout = go.Layout(barmode='group' )
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```

## Categorcial features  ##

```python
catcols_count=defaultdict(list)
for catcol in catcols:
    catcols_count[catcol] = train[catcol].value_counts()
    
def trace_bar(x, y, col_name):
    trace = go.Bar(
    x=x,
    y=y,
    name= col_name)
    return trace

traces = []
for catcol in catcols:
    x = catcols_count[catcol].index.tolist()
    y = catcols_count[catcol].values.tolist()
    traces.append(trace_bar(x, y, catcol))

    
fig = tools.make_subplots(rows =int(len(catcols)/2), cols =2 )



# , subplot_titles= tuple(catcols)
for index, trace in enumerate(traces):
    
    if index >6:
        fig.append_trace(trace, index%7+1,1)
    else:
        fig.append_trace(trace,index%7+1, 2)
  
fig['layout'].update(height=1600, width=1000)
py.plot(fig, filename='customizing-subplot-axes')
```

```python
#find the correlation of interval data(continuous features)
'''
return top correlation list
https://stackoverflow.com/questions/17778394/list-highest-correlation-pairs-from-a-large-correlation-matrix-in-pandas


df = pd.DataFrame(data)
c = df.corr().abs()
s = c.unstack()
so = s.sort_values(kind="quicksort")

'''



def corr_heatmap(v):
    correlations = train[v].corr()
    top_corr = correlations.abs().unstack()
    top_corr = top_corr.sort_values(kind ="quicksort",ascending=False )
    print(top_corr.index)
#     print(top_corr[(top_corr>=0.4) & (top_corr!=1)].index.tolist())
#     print(top_corr[(top_corr>=0.4) & (top_corr!=1)])


    # Create color map ranging between two colors
    cmap = sns.diverging_palette(220, 10, as_cmap=True)

    fig, ax = plt.subplots(figsize=(10,10))
    sns.heatmap(correlations, cmap=cmap, vmax=1.0, center=0, fmt='.2f',
                square=True, linewidths=.5, annot=True, cbar_kws={"shrink": .75})
    plt.show();
    
v = meta[(meta.level == 'interval') & (meta.keep)].index.tolist()
v.append('target')

# print(meta)
# corr_heatmap(v)
# (df.loc[:,['A','B']]) 
# display(train.head())
train_array= np.corrcoef(train[v].values)

print(train_array)
```
